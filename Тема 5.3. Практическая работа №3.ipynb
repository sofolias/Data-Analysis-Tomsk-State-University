### Шаг 1. Создайте программу web-scraping с использованием библиотеки `BeautifulSoup`

```python
# импорт библиотек
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
from lxml import objectify
```

```python
# забираем весь текст с сайта и выводим его через print
url = 'https://quotes.toscrape.com/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
print(soup)
```

<a id=2></a>
### Шаг 2. Реализуйте HTML разбор текста

```python
# найдём все цитаты на странице
quotes = soup.find_all('span', {'class': 'text'})

for quote in quotes:
    print(quote.text)
```

```python
# найдём всех авторов на странице
authors = soup.find_all('small', {'class': 'author'})

for author in authors:
    print(author.text)
```

```python
# найдём все теги на странице
tags = soup.find_all('div', {'class': 'tags'})

for tag in tags:
    print(tag.text.replace('Tags:', '').strip('\n')[26:].splitlines())
```

```python
quotes = soup.find_all('span', {'class': 'text'})
authors = soup.find_all('small', {'class': 'author'})
tags = soup.find_all('div', {'class': 'tags'})

# для вывода датафрейма создаём пустой словарь с соответствующими столбцами: цитата - автор - список тегов
result_output = {'quotes': [], 'authors': [], 'tags': []}

# с помощью цикла перебираем значения и полученное добавляем в словарь
for i in range(0, len(authors)):
    quote = quotes[i].text.strip()
    result_output['quotes'].append(quote)
    print(quote)
    
    author = authors[i].text.strip()
    result_output['authors'].append(author)
    print(author)
    
    tag = tags[i].text.replace('Tags:', '').strip('\n')[3:].split()
    tag_str = ', '.join(tag)  # объединяем значения тегов в строку без квадратных скобок
    result_output['tags'].append(tag_str)
    print(tag_str)
    
    print()
```

### Шаг 3. Выведите результат в DataFrame: Цитата | Автор | Список тегов

```python
# сохраняем полученные результаты в переменную и выводим полученный массив данных
df = pd.DataFrame(data=result_output)
df
```
